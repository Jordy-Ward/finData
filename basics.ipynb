Financial data is usually displayed as time series. As seen in the .csv file in long format. We can read the data doiing the following

```python
#read the .csv file data
prices = pd.read_csv('price_data.csv')
prices
```

Display the data in an easier table format for analysis


```python
print('The csv data in wide format')
prices_wide = prices.pivot(index='date', columns='stock', values='price')
prices_wide
```

Compared to long format which usually has more rows than columns, wide format creates columns for the variable name, in this case the stock name. And we view it by index date, so each row is a date. We can therefore compare each stock price per date visually more easily.
